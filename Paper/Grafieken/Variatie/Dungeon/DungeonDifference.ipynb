```python
# the first line tells Jupyter to place the figures directly in the notebook
%matplotlib inline    
import matplotlib
import numpy as np
from numpy import genfromtxt
import matplotlib.pyplot as plt
```

```python
y1 = genfromtxt('/Users/Robin/Desktop/Grafieken/Dungeon/ans-dun-4.csv', delimiter=';')
y1.sort()
y1 = y1[:len(y1)-11]

y2 = genfromtxt('/Users/Robin/Desktop/Grafieken/Dungeon/ans-dun-5.csv', delimiter=';')
y2.sort()
y2 = y2[:len(y2)-11]

y3 = genfromtxt('/Users/Robin/Desktop/Grafieken/Dungeon/ans-dun-6.csv', delimiter=';')
y3.sort()
y3 = y3[:len(y3)-11]

y4 = genfromtxt('/Users/Robin/Desktop/Grafieken/Dungeon/ans-dun-7.csv', delimiter=';')
y4.sort()
y4 = y4[:len(y4)-11]

y5 = genfromtxt('/Users/Robin/Desktop/Grafieken/Dungeon/prob-dun-4.csv', delimiter=';')
y5.sort()
y5 = y5[:len(y5)-11]

y6 = genfromtxt('/Users/Robin/Desktop/Grafieken/Dungeon/prob-dun-5.csv', delimiter=';')
y6.sort()
y6 = y6[:len(y6)-11]

y7 = genfromtxt('/Users/Robin/Desktop/Grafieken/Dungeon/prob-dun-6.csv', delimiter=';')
y7.sort()
y7 = y7[:len(y7)-11]

y8 = genfromtxt('/Users/Robin/Desktop/Grafieken/Dungeon/prob-dun-7.csv', delimiter=';')
y8.sort()
y8 = y8[:len(y8)-11]

length = len(y8)
x = np.array(range(length))
```

```python
fig = plt.figure(figsize=(7,5))
ax = fig.add_subplot(111)
ax.set_xlabel('Aantal Vergelijkingen', fontsize=12)
ax.set_ylabel('Variatiefactor', fontsize=12)
ax.plot(x,y5,'.',color='0.6',markevery=5, mew=1, ms=6, label="Problog | Dimensie 4")
ax.plot(x,y6,'+',color='0.6',markevery=5,linewidth=3, label="Problog | Dimensie 5")
ax.plot(x,y7,'*',color='0.6',markevery=5,linewidth=3, label="Problog | Dimensie 6")
ax.plot(x,y8,'d',color='0.6',markevery=5,linewidth=3, label="Problog | Dimensie 7")
ax.plot(x,y1,'.',color='black',markevery=5, mew=1, ms=6, label="AnsProlog | Dimensie 4")
ax.plot(x,y2,'+',color='black',markevery=5,linewidth=2, label="AnsProlog | Dimensie 5")
ax.plot(x,y3,'*',color='black',markevery=5,linewidth=2, label="AnsProlog | Dimensie 6")
ax.plot(x,y4,'d',color='black',markevery=5,linewidth=2, label="AnsProlog | Dimensie 7")
plt.xticks(np.arange(min(x), 210, 30))
leg = ax.legend(loc=2,prop={'size':10},fancybox=True)
leg.get_frame().set_alpha(0.5)
plt.savefig('dungeon.pdf')
```
