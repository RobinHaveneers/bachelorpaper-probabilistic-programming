```python
# the first line tells Jupyter to place the figures directly in the notebook
%matplotlib inline    
import matplotlib
import numpy as np
from numpy import genfromtxt
import mpl_toolkits.axisartist as AA

import matplotlib.pyplot as plt
```

```python
y1 = genfromtxt('prob-cm-3-6.csv', delimiter=',')
y1.sort()
y1 = y1[:len(y1)-11]

y2 = genfromtxt('ans-cm-3-6.csv', delimiter=',')
y2.sort()
y2 = y2[:len(y2)-11]

y3 = genfromtxt('prob-cm-4-6.csv', delimiter=',')
y3.sort()
y3 = y3[:len(y3)-11]

y4 = genfromtxt('ans-cm-4-6.csv', delimiter=',')
y4.sort()
y4 = y4[:len(y4)-11]

y5 = genfromtxt('prob-cm-5-4.csv', delimiter=',')
y5.sort()
y5 = y5[:len(y5)-11]

y6 = genfromtxt('ans-cm-5-4.csv', delimiter=',')
y6.sort()
y6 = y6[:len(y6)-11]

length = len(y1)
x = np.array(range(length))
```

```python
fig = plt.figure(figsize=(7,5))
ax = fig.add_subplot(111)


ax.set_xlabel('Aantal Vergelijkingen', fontsize=12)
ax.set_ylabel('Variatiefactor', fontsize=12)
ax.plot(x,y1, '.', color='0.6',markevery=3, mew=1, ms=8, label="Problog | Dimensie 3 - 6 Kleuren")
ax.plot(x,y3, '--', color='0.6',markevery=3,linewidth=2, label="Problog | Dimensie 4 - 6 Kleuren")
ax.plot(x,y5, ':', color='0.6',markevery=3,linewidth=4, label="Problog | Dimensie 5 - 4 Kleuren")
ax.plot(x,y2, '.', color='0',markevery=3,mew=1, ms=8, label="AnsProlog | Dimensie 3 - 6 Kleuren")
ax.plot(x,y4, '--', color='0',markevery=3,linewidth=2,  label="AnsProlog | Dimensie 4 - 6 Kleuren")
ax.plot(x,y6, ':', color='black',markevery=3,linewidth=4, label="AnsProlog | Dimensie 5 - 4 Kleuren")
plt.xticks(np.arange(min(x), 210, 30))
leg = ax.legend(loc=2,prop={'size':11},fancybox=True)
leg.get_frame().set_alpha(0.5)
plt.savefig('chm.pdf')
```
