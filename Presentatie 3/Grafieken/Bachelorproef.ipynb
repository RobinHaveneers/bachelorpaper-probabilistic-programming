```python
# Create a subplot. This gives us more options.
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)   
# This number indicates: number of rows*100 + number of columns*10 + index
#  For more than 10 plots, use a tuple (nr, nc, index)

# Width of one bar.
width = 0.4



x=numpy.array([1,2,3])

y1=numpy.array([2,4,3])
y2=numpy.array([2,4,4])

# Set the plot title
ax.set_title('Hoogste dimensie met gegenereerde totaaloplosing', fontsize=16)

# Set the position of the ticks on the x-axis
ax.set_xticks(x+width)
# Set the labels of the ticks on the x-axis
ax.set_xticklabels(['Chromatic Maze', 'Perfect Maze', 'Dungeons'])

# Remove ticks at the secondary x-axis (at the top)
ax.get_xaxis().tick_bottom()    

# Add some range to the y-axis.
ax.set_ylim((0,max(max(y1), max(y2))+0.5))

# Remove both axis lines
ax.spines["top"].set_visible(False)    
ax.spines["bottom"].set_visible(False)    
ax.spines["right"].set_visible(False)    
ax.spines["left"].set_visible(False)  


# Draw the bars, this returns the rectangles which we need later.
rectangles1 = ax.bar(x, y1, width, color='#00407A', label='ProbLog',edgecolor="none");
rectangles2 = ax.bar(x+width, y2, width, color='#1D8DB0', label='AnsProlog',edgecolor="none");

# Remove the ticks on Y-axis, we'll label the bars
ax.set_yticks([])

# Replace the ticks by labels on the bars themselves
def autolabel(rects):
     # Attach some text labels
     for rect in rects:
         height = rect.get_height()
         ax.text(rect.get_x() + rect.get_width()/2., 1.01*height,
                 '%d' % int(height),
                 ha='center', va='bottom')

autolabel(rectangles1)
autolabel(rectangles2)

# Add a legend, this uses the 'label' arguments of the 'bar' function.
ax.legend(loc=2);

# Save the figure to PDF
fig.savefig('all_generate_dim.pdf')
```
